```python
domain_url='https://www.autostar.com.tw/'
def get_car_data(url,page):
    form_data={    
        'Page': page       
    }
    resp=requests.post(url,data=form_data)
    soup=BeautifulSoup(resp.text,'lxml')
    cars=soup.find_all('div',class_="row item")
    
    datas=[]
    for car in cars:    
        title=car.find('h2').text.strip()
        if '杰運機車' in title:
            continue      
        image_url=domain_url+car.find('img').get('src')
        link=domain_url+car.find('a').get('href')      
        tags=[tag.text.strip() for tag in car.find_all('span',class_="label label-info")]
        infos=[info.text.strip() for info in car.find('div',class_="info").find_all('p')]
        
        datas+=[image_url,title]+tags+infos+[link]
        
    return datas
```

```python
get_car_data(url,1)
```

### 抓取所有車輛

```python
pages=37
datas=[]
for i in range(pages):
    print(f'page:{i+1}')
    datas+=get_car_data(url,i+1)  
    
    
datas
```
